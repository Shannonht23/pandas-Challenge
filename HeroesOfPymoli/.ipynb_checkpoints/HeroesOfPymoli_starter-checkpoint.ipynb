```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
#calculate the number of unique customers 

Players = (len(purchase_data["SN"].unique()))

#create a dataframe

PlayerCountDF = pd.DataFrame({"Total Players":[Players]})

#show the dataframe

PlayerCountDF
      
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Get the number of unique items:
Numberofuniqueitems = len(purchase_data["Item ID"].unique())
print(Numberofuniqueitems)

#get the average price per item:
AveragePrice = purchase_data["Price"].mean()

#get the number of purchases (count)

Purchases = purchase_data["Price"].count()

#get the total of all items purchased(sum)

TotalOfPurchases= purchase_data["Price"].sum()

#create the summary table:
SummaryTableDF= pd.DataFrame({"Number of Unique Items": [Numberofuniqueitems],
                             "Average Price":AveragePrice, 
                             "Number of Purchases":Purchases,
                             "Total Revenue":TotalOfPurchases })

#round the data in the summary table so that decimal amounts have 2 decimal places
#format the Average Item Price to show the dollar sign and 2 decimal places

SummaryTableDF["Average Price"] = SummaryTableDF["Average Price"].map("${:,.2f}".format)


#Format the total of all purchases to show the dollar sign and 2 decimal places

SummaryTableDF["Total Revenue"] = SummaryTableDF["Total Revenue"].map("${:,.2f}".format)
#display the summary table


SummaryTableDF
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#identify the top spending customers(hint) Group by the name column
#calculate the purchase count
Topspenderscount = purchase_data.groupby(["SN"]).count()["Price"]
#calculate the average price per user's purchase
Topspendersavg = purchase_data.groupby(["SN"]).mean()["Price"]

#calculate the total of all of the user's purchases
Topspenderssum = purchase_data.groupby(["SN"]).sum()["Price"]

#make a data frame 
TopspendersDF = pd.DataFrame({
    "Purchase Count": Topspenderscount,
    "Average Purchase Price": Topspendersavg, 
    "Total Purchase Value":Topspenderssum
    
})
#display the data frame and sort the values based on the total of all of the customers purchases:
TopspendersDF = TopspendersDF.sort_values("Total Purchase Value", ascending =False)
#format the data frame dollar columns:
TopspendersDF["Average Purchase Price"] = TopspendersDF["Average Purchase Price"].map("${:,.2f}".format)
TopspendersDF["Total Purchase Value"] = TopspendersDF["Total Purchase Value"].map("${:,.2f}".format)

#display the data frame
TopspendersDF.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Retrieve the Item ID, Item Name, and Item Price columns

ItemData = purchase_data.loc[:,["Item ID","Item Name","Price"]]


#Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value.
ItemCounts = ItemData.groupby(["Item ID","Item Name"]).count()["Price"].rename ("Purchase Count")
Itemavgprice = ItemData.groupby(["Item ID","Item Name"]).mean()["Price"]
Itemtotalprice = ItemData.groupby(["Item ID","Item Name"]).sum()["Price"].rename("Total Purchase Value")


#Create a summary data frame to hold the results
ItemDF=pd.DataFrame({
   "Purchase Count":ItemCounts,
   "Item Price":Itemavgprice,
   "Total Purchase Value":Itemtotalprice})

#Sort the purchase count column in descending order
popularitemDF =ItemDF.sort_values("Purchase Count", ascending=False)

#Optional: give the displayed data cleaner formatting:
popularitemDF["Total Purchase Value"] = popularitemDF["Total Purchase Value"].map("${:,.2f}".format)
popularitemDF["Item Price"] = popularitemDF["Item Price"].map("${:,.2f}".format)


#Display a preview of the summary data frame
popularitemDF.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Sort the above table by total purchase value in descending order:
MostProfitableitemDF =ItemDF.sort_values("Total Purchase Value", ascending=False)
#format the dataframe dollar columns:
MostProfitableitemDF["Item Price"]=MostProfitableitemDF["Item Price"].map("${:,.2f}".format)
MostProfitableitemDF["Total Purchase Value"] = MostProfitableitemDF["Total Purchase Value"].map("${:,.2f}".format)

#Display a preview of the data frame:

MostProfitableitemDF.head()
```
